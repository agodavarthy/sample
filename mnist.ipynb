```python
import torch
import torchvision.datasets as dsets
import numpy as np
```

```python
train_dataset = dsets.MNIST(root='./',
                           train=True,
                           download=True)
```

```python
print len(train_dataset)
```

```python
mnist_numpy_array = np.reshape(train_dataset[0][0], -1, 28*28)
for i in range(1, len(train_dataset)):
    item = np.reshape(train_dataset[0][0], -1, 28*28)
    np.vstack((mnist_numpy_array, item))
```

```python
mnist_numpy_array.shape
```

```python
mnist_norm = mnist_numpy_array/255 # normalized using (x - xmin)/(xmax - xmin)
```

```python
mnist_tensor = torch.from_numpy(mnist_norm)
```
